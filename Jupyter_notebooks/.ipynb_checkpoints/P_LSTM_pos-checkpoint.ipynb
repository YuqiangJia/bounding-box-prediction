```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import import_ipynb
import DataLoader_t
import utils
import time
import numpy as np
```

```python
class args():
    def __init__(self):
        self.base_out_features = 2048
        self.t_obs = 8
        self.batch_size = 20
        self.scene_shape = [3, 144, 256]
        self.base_shape = [2048, 9, 16]
        self.hidden_size = 128
        self.embedding_size = 64
        self.input_size = 2
        self.output_size = 2
        self.sample = False
        self.n_sequences = 20000
        
        #dataset
        self.jaad_dataset = '../../../../data/smailait-data/jaad/annotations'
        self.dtype = 'train'
        self.from_file = True
        self.file = '/data/smailait-data/train_16.csv'
        self.seq_len = 4
        self.predicted_seq_len = 4
        self.crop = 0.3
        self.activity_h = 9
        self.activity_w = 16
        self.truncate = 0
        self.final_frame_offset = 0
        self.loader_workers = 5
        self.loader_shuffle = True
        self.pin_memory = False
        self.image_resize = [144, 256]
        self.image_size = [1080, 1920]
        self.device='cuda'
        
args = args()
```

```python
class P_LSTM(nn.Module):
    def __init__(self):
        super(P_LSTM, self).__init__()
        
        self.encoder = nn.LSTM(input_size=2, hidden_size=128)
        
        self.decoder = nn.LSTMCell(input_size=2, hidden_size=128)
        
        self.out = nn.Linear(in_features=128, out_features=2)
        
        self.activation = nn.ReLU()
        
    def forward(self, input_):
        _, (h, c) = self.encoder(input_.permute(1,0,2))
        outputs = torch.tensor([], device='cuda')
        in_dec = input_[:,-1,:]
        h = h.squeeze(0)
        c = c.squeeze(0)
        for i in range(8):
            h, c = self.decoder(in_dec, (h, c))
            output = self.activation(self.out(h))
            outputs = torch.cat((outputs, output.unsqueeze(1)), dim = 1)
            in_dec = output.detach()
        return outputs
```

```python
net = P_LSTM().to('cuda')
```

```python
train, val = DataLoader_t.data_loader(args)
```

```python
learning_rate = 0.001
optimizer = optim.Adam(net.parameters(), lr=learning_rate)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=0.1)
criterion = nn.MSELoss()
train_scores = []
val_scores = []
```

```python
for epoch in range(100):
    avg_epoch_train_loss = 0
    avg_epoch_val_loss = 0
    ade = 0
    fde = 0
    counter = 0
    for idx, (obs, target) in enumerate(train):
        counter += 1
        obs = obs.type(torch.float32).to(device='cuda')
        target = target.type(torch.float32).to(device='cuda')
        
        net.zero_grad()
        preds = net(obs)
        train_loss = criterion(preds, target)
        train_loss.backward()
        optimizer.step()
        
        avg_epoch_train_loss += float(train_loss)
        
    avg_epoch_train_loss/=counter
    train_scores.append(avg_epoch_train_loss)
    
    counter=0
    for idx, (obs, target) in enumerate(val):
        counter+=1
        obs = obs.type(torch.float32).to(device='cuda')
        target = target.type(torch.float32).to(device='cuda')
        
        with torch.no_grad():
            preds = net(obs)
            val_loss = criterion(preds, target)
            ade += float(utils.ADE_c(preds, target))
            fde += float(utils.FDE_c(preds, target))
            
        avg_epoch_val_loss += float(val_loss)
        
    avg_epoch_val_loss/=counter
    val_scores.append(avg_epoch_val_loss)
    ade/=counter
    fde/=counter
    
    if epoch == 50:
        scheduler.step()

    print('e:', epoch, ' | train_loss:', avg_epoch_train_loss, ' | val_loss:', avg_epoch_val_loss, '| ade:', ade, '| fde:', fde)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.plot(list(range(len(train_scores))), train_scores)
plt.plot(list(range(len(val_scores))), val_scores)
plt.show()
```
