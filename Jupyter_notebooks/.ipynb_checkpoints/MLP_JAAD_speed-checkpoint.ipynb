```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import import_ipynb
import DataLoader_t_speed
import utils
import time
import numpy as np
```

```python
class args():
    def __init__(self):
        self.base_out_features = 2048
        self.t_obs = 8
        self.batch_size = 10
        self.scene_shape = [3, 144, 256]
        self.base_shape = [2048, 9, 16]
        self.hidden_size = 128
        self.embedding_size = 64
        self.input_size = 2
        self.output_size = 2
        self.sample = False
        self.n_sequences = 20000
        
        #dataset
        self.jaad_dataset = '../../../../data/smailait-data/jaad/annotations'
        self.dtype = 'train'
        self.from_file = True
        self.file = '/data/smailait-data/train_crossing_16.csv'
        self.seq_len = 4
        self.predicted_seq_len = 4
        self.crop = 0.3
        self.activity_h = 9
        self.activity_w = 16
        self.truncate = 0
        self.final_frame_offset = 0
        self.loader_workers = 0
        self.loader_shuffle = True
        self.pin_memory = False
        self.image_resize = [144, 256]
        self.image_size = [1080, 1920]
        self.device='cuda'
        
args = args()
```

```python
class MLP_speed(nn.Module):
    def __init__(self):
        super(MLP_speed, self).__init__()
        
        self.in_ = nn.Linear(in_features=2, out_features=64)
        self.hid1 = nn.Linear(in_features=64, out_features=128)
        self.hid2 = nn.Linear(in_features=128, out_features=64)
        self.out_ = nn.Linear(in_features=64, out_features=2)
        self.activation = nn.Hardtanh(min_val=-200, max_val=200)
        
    def forward(self, input_):
        input_ = self.activation(self.in_(input_))
        input_ = self.activation(self.hid1(input_))
        input_ = self.activation(self.hid2(input_))
        output = self.activation(self.out_(input_))
        return output
```

```python
net_speed = MLP_speed().to('cuda')
```

```python
train, val = DataLoader_t_speed.data_loader(args)
```

```python
learning_rate = 0.001
optimizer = optim.Adam(net_speed.parameters(), lr=learning_rate)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=0.1)
criterion = nn.MSELoss()
train_scores = []
val_scores = []
```

```python
for epoch in range(100):
    avg_epoch_train_loss = 0
    avg_epoch_val_loss = 0
    ade = 0
    fde = 0
    counter = 0
    for idx, (obs, target) in enumerate(train):
        counter += 1
        obs = obs.type(torch.float32).to(device='cuda')
        target = target.type(torch.float32).to(device='cuda')
        
        net_speed.zero_grad()
        preds = net_speed(obs)
        train_loss = criterion(preds, target)
        train_loss.backward()
        optimizer.step()
        
        avg_epoch_train_loss += float(train_loss)
        
    avg_epoch_train_loss/=counter
    train_scores.append(avg_epoch_train_loss)
    
    counter=0
    for idx, (obs, target) in enumerate(val):
        counter+=1
        obs = obs.type(torch.float32).to(device='cuda')
        target = target.type(torch.float32).to(device='cuda')
        
        with torch.no_grad():
            preds = net_speed(obs)
            val_loss = criterion(preds, target)
            ade += float(utils.ADE_c(preds, target))
            fde += float(utils.FDE_c(preds, target))
            
        avg_epoch_val_loss += float(val_loss)
        
    avg_epoch_val_loss/=counter
    val_scores.append(avg_epoch_val_loss)
    ade/=counter
    fde/=counter

    print('e:', epoch, ' | train_loss:', avg_epoch_train_loss, ' | val_loss:', avg_epoch_val_loss, '| ade:', ade, '| fde:', fde)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,20))
plt.plot(list(range(len(train_scores))), train_scores)
plt.plot(list(range(len(val_scores))), val_scores)
plt.show()
```
